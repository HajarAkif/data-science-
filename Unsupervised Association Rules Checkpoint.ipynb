```python
dataset = [["Milk", "Onion","Nutmeg","Kidney Beans", "Eggs", "Yogurt"],
            ["Dill", "Onion","Nutmeg","Kidney Beans", "Eggs", "Yogurt"],
            ["Milk", "Apple","Kidney Beans", "Eggs"],
            ["Milk", "Unicorn","Corn","Kidney Beans","Yogurt"],
             ["Corn","Onion","Onion","Kidney Beans","Ice Cream", "Eggs"]  ]
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns = te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
apriori(df, min_support=0.6, use_colnames=True)
```

```python
from mlxtend.frequent_patterns import association_rules
ar = association_rules(apriori(df, min_support=0.6, use_colnames=True),
                 metric = "confidence",
                 min_threshold=0.7)
ar
```

```python
ar[ar['lift']>1]
```

```python
br = association_rules(apriori(df, min_support=0.6, use_colnames=True),
                 metric = "lift",
                 min_threshold=1.01)
br
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv')
df.head()
```

```python
transactions = list()
for i in range(0, 7500):
    transactions.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
trans = [[element for element in transaction if element != 'nan'] for transaction in transactions]
```

```python
te = TransactionEncoder()
trans_array = te.fit(trans).transform(trans)
df_transactions = pd.DataFrame(trans_array, columns = te.columns_)
df_transactions
```

```python
regles = apriori(df_transactions, min_support=0.01, use_colnames=True)
regles
```

```python
regles_association = association_rules(regles,
                 metric = "lift",
                 min_threshold=1.01)
regles_association
```
